```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
%matplotlib inline

mu1 = 0
variance1 = 1
sigma1 = math.sqrt(variance1)
mu2 = 1
variance2 = 2
sigma2 = math.sqrt(variance2)
x = np.linspace(-5,5, 1000)
n_0_1=stats.norm.pdf(x, mu1, sigma1)
n_1_2=stats.norm.pdf(x, mu2, sigma2)
p_x=n_0_1+n_1_2
plt.plot(x, n_0_1)
plt.plot(x, n_1_2)
plt.plot(x, n_0_1/p_x)
plt.plot(x, n_1_2/p_x)
plt.axvline(x=0.840189)
plt.axvline(x=-2.84019)
```
