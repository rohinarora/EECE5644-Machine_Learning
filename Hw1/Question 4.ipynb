```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
%matplotlib inline
txt="Plot of class conditional PDF of 2 classes. Both classes are gaussians. \n                Class 1 is N(0,1) and class 2 is N(1,2)"
fig = plt.figure(figsize=(10,10));
fig.text(.15,0.025,txt,fontsize=15);

plt.ylim(0,1)
plt.xlim(-4.5,4.5)
mu1 = 0
variance1 = 1
sigma1 = math.sqrt(variance1)
mu2 = 1
variance2 = 2
sigma2 = math.sqrt(variance2)
x = np.linspace(-6,6, 1000)
n_0_1=stats.norm.pdf(x, mu1, sigma1)
n_1_2=stats.norm.pdf(x, mu2, sigma2)
p_x=n_0_1+n_1_2
plt.plot(x, n_0_1,label="class conditonal of class 1")
plt.plot(x, n_1_2,label="class conditonal of class 2");
plt.axvline(x=0.840189,label="decision boundary x=−2.84019",color='black',ymax=0.45,linestyle='dashed')
plt.axvline(x=-2.84019,label="decision boundary x=0.840189",color='black',ymax=0.45,linestyle='dashed');
plt.axvspan( -2.84019, 0.840189, alpha=0.2, color='red',ymax=.45,label='decide class 2')
plt.axvspan( -10, -2.84019, alpha=0.1, color='green',ymax=.45)
plt.axvspan( 0.840189, 10, alpha=0.1, color='green',ymax=.45,label='decide class 1')
plt.legend();

plt.title('Plot of class conditional PDF of 2 classes',fontsize=12)
plt.xlabel('Random Variable X',fontsize=12)
plt.ylabel('Probability',fontsize=12);
```

```python
txt="           Plot of posterior PDF of 2 classes.\nThe posteriors sum to 1 as shown by the green line"
fig = plt.figure(figsize=(10,10));
fig.text(.25,0.025,txt,fontsize=15);
plt.ylim(0,1.5)
plt.xlim(-6,6)
plt.plot(x, n_0_1/p_x,label="posterior of class 1")
plt.plot(x, n_1_2/p_x,label="posterior of class 2")
plt.plot(x, n_1_2/p_x+n_0_1/p_x,label="sum of posteriors")
plt.axvline(x=0.840189,label="decision boundary x=−2.84019",color='black',ymax=0.8,linestyle='dashed')
plt.axvline(x=-2.84019,label="decision boundary x=0.840189",color='black',ymax=0.8,linestyle='dashed');
plt.axvspan( -2.84019, 0.840189, alpha=0.2, color='red',ymax=.8,label='decide class 2')
plt.axvspan( -10, -2.84019, alpha=0.1, color='green',ymax=.8)
plt.axvspan( 0.840189, 10, alpha=0.1, color='green',ymax=.8,label='decide class 1')


plt.legend();
plt.title('Plot of posterior PDF of 2 classe',fontsize=12)
plt.xlabel('Random Variable X',fontsize=12)
plt.ylabel('Probability',fontsize=12);
```

```python
txt="Posteriors and class conditonal PDFs of 2 classes together"
fig = plt.figure(figsize=(10,10));
fig.text(.35,0.025,txt,fontsize=15);
plt.ylim(0,1.5)
plt.xlim(-6,6)
plt.plot(x, n_0_1,label="class conditonal of class 1")
plt.plot(x, n_1_2,label="class conditonal of class 2");
plt.plot(x, n_0_1/p_x,label="posterior of class 1")
plt.plot(x, n_1_2/p_x,label="posterior of class 2")
plt.axvline(x=0.840189,label="decision boundary x=−2.84019",color='black',ymax=0.8,linestyle='dashed')
plt.axvline(x=-2.84019,label="decision boundary x=0.840189",color='black',ymax=0.8,linestyle='dashed');
plt.axvspan( -2.84019, 0.840189, alpha=0.2, color='red',ymax=.8,label='decide class 2')
plt.axvspan( -10, -2.84019, alpha=0.1, color='green',ymax=.8)
plt.axvspan( 0.840189, 10, alpha=0.1, color='green',ymax=.8,label='decide class 1')

plt.legend();
plt.title('Posteriors and class conditonal PDFs',fontsize=12)
plt.xlabel('Random Variable X',fontsize=12)
plt.ylabel('Probability',fontsize=12);
```

### Q4 Part 4 Error as sigma >> 1

```python
'''
Here i measure one of the terms of P(error) - P(error) of misclassifying class 1- P(D=1|L=2)
We decide class 1 in the region-
−√2 lnσ < x <√2 lnσ

P(D=1|L=2)= integral in above region of (class conditional PDF of class 2). 

Conditional PDF here is gaussian. This integral has been measured by -
CDF (√2 lnσ) -  CDF (−√2 lnσ)

The class conditional PDF of class 2 itself varies with σ, and it has been taken care


Result- We see P(error) of misclassifying class 1 goes dramtically down as sigma increases. 
'''

for sigma in range(2,100,10):
    f_sigma=scipy.stats.norm(0, sigma).cdf
    print (f_sigma(np.sqrt(2*np.log(sigma)))-f_sigma(-np.sqrt(2*np.log(sigma))))
```
